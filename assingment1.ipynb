**Import required libraries and Load Data**

**Binary Conversion**

```python
# Converting the categori'cal variable into numerical
varlist =  [ 'Gender']

# Defining the map function
def binary_map(x):
    return x.map({'Male': 1, "Female": 0})

# Applying the function to the Social_Network_Ads list
dataset[varlist] = dataset[varlist].apply(binary_map)

# Check the Social_Network_Ads dataframe now
dataset
```

**Categorical Conversion**

```python
# Dropping the first column from status dataset
status = pd.get_dummies(dataset['Social Media'], drop_first = True)

# Adding the status to the original dataset dataframe
dataset = pd.concat([dataset, status], axis = 1)

# Dropping 'social media' as we have created the dummies for it
dataset.drop(['Social Media'], axis = 1, inplace = True)

dataset
```

**Outliers Detection**

```python
def detect_outliers_zscore(data):
    outliers = []
    thres = 3
    mean = np.mean(data)
    std = np.std(data)
    #print(mean, std)
    for i in data:
        #print(i)
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers# Driver code
```

```python
sample_outliers_Gender = detect_outliers_zscore(dataset['Gender'])
sample_outliers_EstimatedSalary = detect_outliers_zscore(dataset['EstimatedSalary'])
sample_outliers_Age = detect_outliers_zscore(dataset['Age'])
print("Outliers in Gender column from Z-scores method: ", sample_outliers_Gender)
print("Outliers in estimatedsalary column from Z-scores method: ", sample_outliers_EstimatedSalary)
print("Outliers in Age column from Z-scores method: ", sample_outliers_Age)
```

```python
from scipy.stats import zscore
z1 = np.abs(zscore(dataset['EstimatedSalary']))
z2 = np.abs(zscore(dataset['Age']))
z3 = np.abs(zscore(dataset['Gender']))
dataset['Zscore_EstimatedSalary'] = z1
dataset['Zscore_Age'] = z2
dataset['Zscore_Gender'] = z3
outliers_price=len(dataset[dataset['Zscore_EstimatedSalary']>3])
outliers_area=len(dataset[dataset['Zscore_Age']>3])
outliers_bed=len(dataset[dataset['Zscore_Gender']>3])
data_z = dataset[dataset['Zscore_EstimatedSalary']<3]
data_z = dataset[dataset['Zscore_Age']<3]
data_z = dataset[dataset['Zscore_Gender']<3]
dataset=data_z.drop(['Zscore_EstimatedSalary', 'Zscore_Age', 'Zscore_Gender'], axis=1)
dataset
```
